```python
def distancia_manhattan(x1,y1,x2,y2):

    distancia = abs(x1-x2) + abs(y1-y2)

    return distancia

# coordenadas
inicio = (1,2)
objetivo = (4,6)

resultado = distancia_manhattan(inicio[0],inicio[1],objetivo[0],objetivo[1])

print("Distancia Manhattan:",resultado)
```
