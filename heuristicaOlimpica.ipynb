```python
def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

lugares = {
    "Entrada": (0,0),
    "Frutas": (1,1),
    "Aseo": (2,2),
    "Caja": (3,3)
}

inicio = "Entrada"
objetivo = "Caja"

recorrido = ["Entrada", "Frutas", "Aseo", "Caja"]

print("Recorrido:")
for lugar in recorrido:
    print(lugar, "-> Heurística:", manhattan(lugares[lugar], lugares[objetivo]))
```
