```python
olimpica = "promocion"
competencia = "no promocion"

print("Olímpica:", olimpica)
print("Competencia:", competencia)

if olimpica == "promocion" and competencia == "no promocion":
    print("Resultado: Olímpica gana más clientes")

elif olimpica == competencia:
    print("Resultado: Empate")

else:
    print("Resultado: Competencia gana")
```

```python
precio_olimpica = 8000
precio_otra = 10000

print("Precio en Olímpica:", precio_olimpica)
print("Precio en otra tienda:", precio_otra)

if precio_olimpica < precio_otra:
    print("El cliente elige Olímpica")
else:
    print("El cliente elige otra tienda")
```
