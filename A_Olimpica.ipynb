```python
import heapq
import time

grafo = {
    'Entrada': {'Frutas': 1, 'Lacteos': 3},
    'Frutas': {'Carnes': 1, 'Aseo': 4},
    'Lacteos': {'Bebidas': 2},
    'Carnes': {'Caja': 5},
    'Aseo': {'Caja': 1},
    'Bebidas': {'Caja': 2},
    'Caja': {}
}

heuristica = {
    'Entrada': 7,
    'Frutas': 6,
    'Lacteos': 4,
    'Carnes': 5,
    'Aseo': 2,
    'Bebidas': 1,
    'Caja': 0
}

def a_estrella(inicio, objetivo):

    cola = []
    heapq.heappush(cola, (0, inicio))

    camino = {}
    costo = {inicio: 0}

    visitados = []

    while cola:
        actual = heapq.heappop(cola)[1]

        # Mostrar recorrido
        print("Visitando:", actual)
        visitados.append(actual)
        time.sleep(1)

        if actual == objetivo:
            break

        for vecino, peso in grafo[actual].items():
            nuevo_costo = costo[actual] + peso

            if vecino not in costo or nuevo_costo < costo[vecino]:
                costo[vecino] = nuevo_costo
                prioridad = nuevo_costo + heuristica[vecino]
                heapq.heappush(cola, (prioridad, vecino))
                camino[vecino] = actual

    # reconstruir camino final
    ruta = []
    nodo = objetivo

    while nodo != inicio:
        ruta.append(nodo)
        nodo = camino[nodo]

    ruta.append(inicio)
    ruta.reverse()

    return ruta

# Ejecutar
resultado = a_estrella('Entrada', 'Caja')

print("\nCamino final encontrado:")
print(" -> ".join(resultado))
```
