```python
def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

lugares = {
    "Bebidas": (0,2),
    "Lacteos": (1,2),
    "Aseo": (2,2),
    "Caja": (3,2)
}

inicio = "Bebidas"
objetivo = "Caja"

recorrido = ["Bebidas", "Lacteos", "Aseo", "Caja"]

print("Recorrido:")
for lugar in recorrido:
    print(lugar, "-> Heurística:", manhattan(lugares[lugar], lugares[objetivo]))
```
